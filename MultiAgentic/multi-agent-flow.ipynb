```python
from crewai import Agent, Task, Crew
import ollama
import pkg_resources

# Get the version of CrewAI

crewai_version = pkg_resources.get_distribution("crewai").version

print(crewai_version)
```

```python
event_planning_agent = Agent(
    role="Event Planning Agent",
    goal="Responsible for creating event plans, schedules, and itineraries for this event {topic}",
    backstory="You're working as a Event Planning Manager at a Event Management Company."
              "You have to understand client requirements, allocate resources, and generate detailed timelines and "
              "You are responsible for planning well-organized events "
              "about the topic: {topic}.",

    allow_delegation=False,
    llm='ollama/bigllama/mistralv01-7b',
	verbose=True,
    memory = True
)

vendor_coordination_agent = Agent(
    role="Vendor Co-Ordination Agent",
    goal="Manages communication and coordination with vendors (e.g., catering, decorations, audio-visual equipment). "
         " about the event: {topic}",
    backstory="You are a Vendor Corordination Agent who received a event planning " 
               "from the Event Planning Manager. "
              "Your goal is to streamline vendor interactions," 
              "and come up with the final vendor list for the event.",
    llm='ollama/bigllama/mistralv01-7b',
    verbose=True,
    memory = True
)

budget_management_agent = Agent(
    role="Budget Management Agent",
    goal="Manage budget of the given event plan {topic} to " 
        "maximize cost savings without compromising event quality or client satisfaction.",
    backstory="You are an Budget Management Agent who receives event plans and vendor quotes "
              "from the Event Planning Manager and Vendor Coordination Agent. "
              "Your goal is to review the event planning and the vendor list"
              "and come up with the final budget for the event.",
    allow_delegation=False,
    llm='ollama/bigllama/mistralv01-7b',
    verbose=True,
    memory=True
)


```

```python
event_planning = Task(
    description=(
        "1. Analyze client requirements for {topic} event including theme, size, and location.\n"
        "2. Create detailed timeline with milestones from pre-event prep to post-event wrap-up.\n"
        "3. Allocate human/resources (staff, equipment, venues) using historical event data.\n"
        "4. Generate risk mitigation plan for potential disruptions."
    ),
    expected_output="Comprehensive event plan document with:\n"
                   "- Hour-by-hour schedule\n"
                   "- Resource allocation matrix\n"
                   "- Client approval-ready presentation",
    agent=event_planning_agent,
)

vendor_coordination = Task(
    description=(
        '''1. Identify required vendors for {topic} event (catering, AV, security, etc.)\n"
        "2. Negotiate contracts with 3+ vendors per category for best pricing/service\n"'''
    ),
    expected_output="Vendor coordination report containing:\n"
                   "- Signed contracts with key terms highlighted\n"
                   "- Contact escalation matrix\n"
                   "- Delivery timelines mapped to event schedule\n"
                   "- Backup vendor list",
    agent=vendor_coordination_agent,
)

budget_management = Task(
    description=(
        '''1. Analyze cost estimates from event plan and vendor quotes\n"
        "2. Allocate budget to categories with 15% contingency reserve\n"
        "4. Identify 3-5 cost-optimization opportunities without quality impact'''
    ),
    expected_output="Budget management package with:\n"
                   "- Approved budget breakdown\n"
                   "- Cost-saving recommendations\n"
                   "- ROI projection report",
    agent=budget_management_agent
)
```

```python
crew = Crew(
    agents=[event_planning_agent, vendor_coordination_agent, budget_management_agent],
    tasks=[event_planning, vendor_coordination, budget_management]

)
```

```python
result = crew.kickoff(inputs={"topic": "Birthday party for a 5-year-old for 100 people with return gift and buffet within 1 lakh budget."})
```
